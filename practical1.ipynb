```python
# calc
a=int(input("num1: "))
b=int(input("num2: "))
sum=a+b
diff=a-b
mul=a*b
div=a/b
print("1: sum, 2: difference, 3: multiplication, 4: division")
ch=int(input("enter your choice: "))
if ch==1:
  print(sum)
elif ch==2:
  print(diff)
elif ch==3:
  print(mul)
elif ch==4:
  print(div)
else:
  print("invalid input")
```

```python
# average of marks
st=10
sum=0
for i in range(st):
  m=float(input("enter marks: "))
  sum+=m
avg=sum/st
print("sum: ", sum)
print("average marks: ", avg)
```
